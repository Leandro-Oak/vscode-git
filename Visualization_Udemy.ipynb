## Keyword arguments

```python




#lmplot is a function from Seaborn used to plot linear relationships (scatter plots + optional regression lines) between two variables.

# Key parameters:

# x, y: Columns to plot.
# demographics: demographicsFrame.
# fit_reg: Show regression line (True/False).
# hue: Color-code by category.
# col / row: Create subplots by category.
# height: Size of each subplot (in inches).
# scatter_kws: Dict to customize markers (e.g., size, transparency).


#change the marker color and size with scatter_kws={"s":200}
vis4 = sns.lmplot(
    x='InternetUsers', 
    y='BirthRate', 
    data=demographics, 
    fit_reg=True,
    hue='IncomeGroup', 
    height=8,
    scatter_kws={"s":100})
```

## Stacked Distribution / Histograms

```python
#version 1
#Graph with filters of two columns
#And show them on top of each other


plt.hist(movies[movies['Genre']=="Action"].BudgetMillions,bins=15 )
plt.hist(movies[movies['Genre']=="Drama"].BudgetMillions,bins=15)
plt.hist(movies[movies['Genre']=="Romance"].BudgetMillions,bins=15)
plt.show()
```

```python
#Version 2
#Side side of each other

plt.hist([movies[movies['Genre']=="Action"].BudgetMillions,\
         movies[movies['Genre']=="Drama"].BudgetMillions,\
         movies[movies['Genre']=="Romance"].BudgetMillions],bins=15)
plt.show()
```

```python
#Version 3
#Stacked on top of each other


plt.hist([movies[movies['Genre']=="Action"].BudgetMillions,\
         movies[movies['Genre']=="Drama"].BudgetMillions],bins=15, stacked = True,)
plt.show()
```

```python
#Version 4
# Using For looping to define multiple Genre
filter= []
mylabels = []
for x in movies['Genre'].cat.categories:
  filter.append(movies[movies['Genre']== x].BudgetMillions)
  mylabels.append(x)


h = plt.hist(filter, bins=30, stacked=True, rwidth=3, label=mylabels)
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# Simulated Ontario cities with coordinates
locations = [
    {"city": "Toronto", "lat": 43.651070, "lon": -79.347015},
    {"city": "Ottawa", "lat": 45.421530, "lon": -75.697193},
    {"city": "Mississauga", "lat": 43.589045, "lon": -79.644120},
    {"city": "Hamilton", "lat": 43.255721, "lon": -79.871102},
    {"city": "London", "lat": 42.984923, "lon": -81.245277},
    {"city": "Kitchener", "lat": 43.451639, "lon": -80.492533},
    {"city": "Windsor", "lat": 42.314937, "lon": -83.036363},
    {"city": "Sudbury", "lat": 46.491204, "lon": -80.990005},
    {"city": "Thunder Bay", "lat": 48.380895, "lon": -89.247682},
    {"city": "Kingston", "lat": 44.231172, "lon": -76.485954}
]

project_types = ["Residential", "Commercial", "Infrastructure", "Public Facility"]

def generate_project(i):
    loc = random.choice(locations)
    start_date = datetime(2023, 1, 1) + timedelta(days=random.randint(0, 500))
    duration = random.randint(60, 360)
    end_date = start_date + timedelta(days=duration)
    
    return {
        "project_id": f"P-{i:04d}",
        "city": loc["city"],
        "latitude": loc["lat"] + np.random.normal(0, 0.01),
        "longitude": loc["lon"] + np.random.normal(0, 0.01),
        "project_type": random.choice(project_types),
        "estimated_cost_million": round(np.random.uniform(1, 100), 2),
        "start_date": start_date.date(),
        "end_date": end_date.date(),
        "status": random.choice(["Planned", "In Progress", "Completed"])
    }

# Create DataFrame
df_projects = pd.DataFrame([generate_project(i) for i in range(1, 101)])
```
