```python
import matplotlib.pyplot as plt
import seaborn as sns

import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'DejaVu Sans'

# Create a 2x4 grid now
f, axes = plt.subplots(2, 4, figsize=(30, 15))

# KDE plots
sns.kdeplot(
    x=movies["BudgetMillions"], 
    y=movies["AudienceRatings"], 
    ax=axes[0, 0], 
    fill=True
)

sns.kdeplot(
    x=movies["BudgetMillions"], 
    y=movies["CriticRatings"], 
    ax=axes[0, 1], 
    fill=True
)

sns.scatterplot(
    data=movies, 
    x='CriticRatings', 
    y='AudienceRatings', 
    hue='Genre', 
    ax=axes[0, 2]
)

# --- Scorecards go here ---
score_ax = axes[0, 3]
score_ax.axis('off')  # Hide axes

# Calculate metrics
corr_budget_critic = movies["BudgetMillions"].corr(movies["CriticRatings"])
corr_budget_audience = movies["BudgetMillions"].corr(movies["AudienceRatings"])
mean_budget = movies["BudgetMillions"].mean()
mean_critic = movies["CriticRatings"].mean()
mean_audience = movies["AudienceRatings"].mean()

# Add scorecard text
score_ax.text(0.1, 0.9, f"📊 Corr(Budget, Critic)\n{corr_budget_critic:.2f}", fontsize=16, bbox=dict(facecolor='lightblue', alpha=0.6))
score_ax.text(0.1, 0.7, f"🎯 Corr(Budget, Audience)\n{corr_budget_audience:.2f}", fontsize=16, bbox=dict(facecolor='lightgreen', alpha=0.6))
score_ax.text(0.1, 0.5, f"💰 Avg Budget\n{mean_budget:.1f}M", fontsize=16, bbox=dict(facecolor='lightgray', alpha=0.6))
score_ax.text(0.1, 0.3, f"⭐️ Avg Critic Rating\n{mean_critic:.1f}", fontsize=16, bbox=dict(facecolor='lavender', alpha=0.6))
score_ax.text(0.1, 0.1, f"👥 Avg Audience Rating\n{mean_audience:.1f}", fontsize=16, bbox=dict(facecolor='mistyrose', alpha=0.6))

# Lower row plots
sns.barplot(
    data=movies, 
    x='Year', 
    y='BudgetMillions', 
    ax=axes[1, 0]
)

sns.boxplot(
    data=movies, 
    x="Genre", 
    y="CriticRatings", 
    ax=axes[1, 1]
)

sns.violinplot(
    data=movies, 
    x="Genre", 
    y="CriticRatings", 
    ax=axes[1, 2]
)

# Last slot (optional: another scorecard, or leave blank)
axes[1, 3].axis('off')  # Empty or reuse for logo/title

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# Simulated Ontario cities with coordinates
locations = [
    {"city": "Toronto", "lat": 43.651070, "lon": -79.347015},
    {"city": "Ottawa", "lat": 45.421530, "lon": -75.697193},
    {"city": "Mississauga", "lat": 43.589045, "lon": -79.644120},
    {"city": "Hamilton", "lat": 43.255721, "lon": -79.871102},
    {"city": "London", "lat": 42.984923, "lon": -81.245277},
    {"city": "Kitchener", "lat": 43.451639, "lon": -80.492533},
    {"city": "Windsor", "lat": 42.314937, "lon": -83.036363},
    {"city": "Sudbury", "lat": 46.491204, "lon": -80.990005},
    {"city": "Thunder Bay", "lat": 48.380895, "lon": -89.247682},
    {"city": "Kingston", "lat": 44.231172, "lon": -76.485954}
]

project_types = ["Residential", "Commercial", "Infrastructure", "Public Facility"]

def generate_project(i):
    loc = random.choice(locations)
    start_date = datetime(2023, 1, 1) + timedelta(days=random.randint(0, 500))
    duration = random.randint(60, 360)
    end_date = start_date + timedelta(days=duration)
    
    return {
        "project_id": f"P-{i:04d}",
        "city": loc["city"],
        "latitude": loc["lat"] + np.random.normal(0, 0.01),
        "longitude": loc["lon"] + np.random.normal(0, 0.01),
        "project_type": random.choice(project_types),
        "estimated_cost_million": round(np.random.uniform(1, 100), 2),
        "start_date": start_date.date(),
        "end_date": end_date.date(),
        "status": random.choice(["Planned", "In Progress", "Completed"])
    }

# Create DataFrame
df_projects = pd.DataFrame([generate_project(i) for i in range(1, 101)])
```

```python
df_projects
```

```python
import folium

# Center of Ontario (roughly)
map_center = [44.0, -79.5]

# Create base map
m = folium.Map(location=map_center, zoom_start=6, tiles="CartoDB positron")

# Add project markers
for _, row in df_projects.iterrows():
    popup = f"""
    <b>Project ID:</b> {row['project_id']}<br>
    <b>City:</b> {row['city']}<br>
    <b>Type:</b> {row['project_type']}<br>
    <b>Cost:</b> ${row['estimated_cost_million']}M<br>
    <b>Status:</b> {row['status']}<br>
    <b>Start:</b> {row['start_date']}<br>
    <b>End:</b> {row['end_date']}
    """
    color = {
        "Planned": "blue",
        "In Progress": "orange",
        "Completed": "green"
    }.get(row['status'], "gray")
    
    folium.CircleMarker(
        location=[row['latitude'], row['longitude']],
        radius=6,
        color=color,
        fill=True,
        fill_color=color,
        fill_opacity=0.7,
        popup=folium.Popup(popup, max_width=300)
    ).add_to(m)
```
